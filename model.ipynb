### Обучение модели sBERT

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
```

### Получение эмбеддингов 

```python
resume = "Опыт работы с Python, анализ данных, машинное обучение."
job = "Требуется специалист по машинному обучению, Python, анализ данных."

emb_resume = model.encode(resume, convert_to_tensor=True)
emb_vacancy = model.encode(job, convert_to_tensor=True)
```

```python
similarity = util.pytorch_cos_sim(emb_resume, emb_vacancy)
print(f'Схожесть: {similarity.item():.4f}')
```
